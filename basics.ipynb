```python
def my_decorator(func):
    def wrapper(*args):
        print("Before the function")
        func(*args)
        print("After the function")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hi")
```

```python
@my_decorator
def add(a,b):
    print(a+b)
```

```python
def search_by_name(file_path, name_to_search):
    found = False  # Flag to track if the name is found

    with open(file_path, "r") as f:
        for line in f:
            id, name, age = line.strip().split(",")
            if name.strip() == name_to_search:
                print({"id": int(id), "name": name.strip(), "age": int(age)})
                found = True  # Set flag to True if name is found

    if not found:
        print(f"No records found for the name: {name_to_search}")

search_by_name("test.txt", input("enter search name"))
```

```python

def add_task():
    task = input("Enter a task: ")
    with open("todo.txt", "r+") as f:
        lines = f.readlines()
        task_id = len(lines) + 1
        f.write(str(task_id) + "," + task + "\n")
    

def view_tasks():
    try:
        with open("todo.txt", "r") as f:
            tasks = f.readlines()
            if tasks:
                print("\nTasks:")
                for task in tasks:
                    print("- " + task.strip())
            else:
                print("No tasks found.\n")
    except FileNotFoundError:
        print("No tasks file found.\n")

while True:
  
    choice = input("Enter input: 1. Add Task,  2. View Tasks, 0: Exit")

    if choice == "1":
        add_task()
        break
    elif choice == "2":
        view_tasks()
        break
    else:
        print("Invalid choice.\n")
        break
```

```python
def is_palindrome(number):
    num_str = str(number)
    return num_str == num_str[::-1]

n = int(input("Enter a number: "))
if is_palindrome(n):
    print("It's a palindrome!")
else:
    print("Not a palindrome.")
```
